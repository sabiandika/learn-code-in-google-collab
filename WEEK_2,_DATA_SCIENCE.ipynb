DAY 2. NUMPY ADVANCED


Broadcasting di NumPy adalah fitur yang memungkinkan array dengan ukuran berbeda tetap bisa dilakukan operasi matematika tanpa harus menyamakan ukuran secara manual


Rules broadcasting di NumPy ada 3 aturan utama:

Dimensi dibandingkan dari kanan ke kiri
Shape array dicek mulai dari dimensi paling belakang.
Dua dimensi cocok jika:
ukurannya sama
contoh: (3,4) dengan (3,4)
atau salah satunya bernilai 1
contoh: (3,4) dengan (1,4)
Jika tidak sama dan tidak ada yang 1

```python
import numpy as np

arr = np.array([1,2,3,5])
scalar = 2

result = arr + scalar
print(result)

matrix = np.array([[1,2,3],[4,5,6]])
scalar = 2
result = matrix + scalar
print("\n",result)
```

Aggregation di NumPy adalah proses menggabungkan banyak nilai menjadi satu nilai hasil.

Biasanya dipakai untuk:

jumlah (sum)
rata-rata (mean)
nilai terbesar (max)
nilai terkecil (min)

```python

arr = np.array([[1,2,3], [4,5,6]])

print("sum",np.sum(arr))
print("mean",np.mean(arr))
print("max",np.max(arr))
print("min",np.min(arr))
print("Standar deviasi",np.std(arr))
print("Sum Baris:",np.sum(arr,axis=1))
print("Sum Kolom:",np.sum(arr,axis=0))

```

Boolean indexing dan filtering di NumPy digunakan untuk mengambil data berdasarkan kondisi (True / False).

```python
a = np.array([1,2,3,4,5,6])
print(a > 3,"\n")

modulus = a[a % 2 == 0]
print(modulus,"\n")

a[a > 4] = 0
print("Modified Array",a)
```

Random number generation di NumPy digunakan untuk membuat angka acak.

seed() dipakai supaya angka random yang dihasilkan selalu sama setiap program dijalankan.

random = membuat angka acak
seed = mengunci hasil random agar konsisten/reproducible.

```python

import numpy as np
random_array = np.random.rand(3,3)
print ("random array: \n ",random_array)

random_int = np.random.randint(0,10,size=(2,3))
print("random integer: \n",random_int)

np.random.seed(42)
random_seed = np.random.rand(3)
print("random seed: \n",random_seed)
```

filter random dataset

```python
import numpy as np

dataset = np.random.randint(1,50, size =(5,5))
print("original random: \n", dataset)

dataset[dataset < 20] = 0
print("\n",dataset)

print("sum: ", np.sum(dataset))
print("mean: ", np.mean(dataset))
print("max :", np.max(dataset))
print("min :",np.min(dataset))
print("standar deviasi: ", np.std(dataset))
```
