```python
#app penampil resep
def buka_recipes(file_path):
  try:
    with open(file_path,"r") as file:
      content = file.read()
      recipes = content.split("\n\n")
      recipe_dict = {}
      for recipe in recipes:
        lines = recipe.split("\n")
        if len(lines) >= 3: # Ensure there are enough lines for name, ingredients, and instructions
          name = lines[0].strip()
          ingredients = lines[1].replace('Ingredients: ','').strip()
          instructions = lines[2].replace('Instructions: ','').strip()
          recipe_dict[name] = {"ingredients": ingredients, "instructions": instructions}
      return recipe_dict
  except FileNotFoundError:
    print(f"File {file_path} tidak ditemukan.")
    return {}

#step 2
def show_recipes():
  print("\nAplikasi Penyimpanan Resep Makanan")
  print("==================================")
  print("1. lihat semua resep")
  print("2. cari resep berdasarkan kata kunci")
  print("3. keluar")

#step 3
def view_recipes(recipes):
  name = input("Masukin Nama Makanan: ").strip()
  if name in recipes:
    print(f"Resep: {name}")
    print(f"Bahan-Bahan: {recipes[name]['ingredients']}")
    print(f"Instruksi: {recipes[name]['instructions']}")
  else:
    print(f"Resep {name} tidak ditemukan.")

#step 4
recipe_file = "recipes.txt"
recipes = buka_recipes(recipe_file)

#step 5
while True:
  show_recipes()
  pilihan = input("pilih menu (1/2/3): ")
  if pilihan == "1":
    print("Lihat Semua Resep Makanan")
    for name in recipes:
      print(name)
  elif pilihan == "2":
    view_recipes(recipes)
  elif pilihan == "3":
    print("Terima kasih telah menggunakan aplikasi ini!")
    break
  else:
    print("Pilihan tidak valid. Silakan coba lagi.")
```

```python
def write_file(filename, items):
    with open(filename, "w") as file:
      for item in items:
        file.write(item + "\n")

def read_file(filename):
  try:
    with open(filename, "r") as file:
      items = file.readlines()
      print("Items in the file: ")
      for item in items:
        print(item.strip())
  except FileNotFoundError:
    print(f"File {filename} tidak ada")

  fruits = ["apple", "banana", "cherry"]
  write_file("fruits.txt", fruits)
  read_file("fruits.txt")
```

map(),filter(),reduce()

1. map() → Transform Data

Digunakan buat mengubah setiap elemen dalam iterable.

2. filter() → Menyaring Data

Digunakan buat ambil elemen yang memenuhi kondisi

3. reduce() → Menggabungkan Data

Digunakan buat menggabungkan semua elemen jadi satu nilai

```python
from functools import reduce
angka = [1, 2, 3, 4, 5]

#map
hasil = list(map(lambda x: x * 2, angka))
print("map: ", hasil)

#filter
dua = list(filter(lambda x: x % 2 == 0, angka))
print("filter: ", dua)

#reduce
tiga = reduce(lambda x, y : x * y, angka)
print("reduce: ", tiga)
```
